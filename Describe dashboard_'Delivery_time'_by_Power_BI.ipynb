### **1. Phân tích tình hình vận hành**

Bộ dữ liệu cung cấp thông tin về tình hình vận hành của các đơn hàng từ khi được mua bởi khách hàng trên trang thương mại điện tử, trải qua quá trình vận chuyển và đưa tới tay khách hàng. 
Chúng ta sẽ xem xét tình hình vận chuyển bằng thời gian giao hàng trung bình (Average of Total Delivery Time) theo các chiều dữ liệu khác nhau.

Nhìn chung trong năm 2017, tổng thời gian giao hàng trung bình là 310.56 Giờ (~ 13 ngày). Thời gian xác nhận đơn hàng là 10.39 giờ.
Đến năm 2018, tổng thời gian giao hàng trung bình đã được rút ngắn lại còn 289.16 Giờ (~ 12 ngày). Thời gian xác nhận đơn hàng là 10.68 giờ

Để cải thiện về thời gian giao hàng, chúng ta cần tìm ra các yếu tố ảnh hưởng đến thời gian giao hàng của công ty như vị trí địa lý của khách hàng, các loại sản phẩm, kích thước sản phẩm hay nhà vận chuyển,.. để ra quyết định chỉnh sửa, nâng cao chất lượng hoặc loại bỏ chính xác.

image.png



Quá trình vận chuyển được chia ra làm 3 giai đoạn, lần lượt là Approving Time: Thời gian từ lúc khách hàng đặt hàng cho đến khi đơn được xác nhận; Delivering to Carrier Time: Thời gian người bán giao cho đơn vị vận chuyển; Delivering to Customer Time: Thời gian đơn vị vận chuyển hoàn tất đơn hàng đến cho khách hàng.

Ở pie chart này, có thể thấy phần chiếm nhiều thời gian nhất là giai đoạn đơn vị vận chuyển lấy hàng từ người bán và chuyển đến tay khách hàng (73.6% tổng thời gian trung bình). Cho nên, chúng ta cần tập trung cải thiện bước này trong quá trình vận chuyển

image.png

Trên thực tế, các sản phẩm được bán trên sàn thương mại điện tử rất đa dạng về kích thước, cân nặng. Khi đó, hình thức vận chuyển có thể sẽ khác nhau với từng sản phẩm ảnh hưởng trực tiếp đến thời gian giao hàng. Hai biểu đồ trên chỉ ra tổng thời gian giao hàng giảm xuống đáng kể xuống chỉ còn 100-150 giờ (4-6 ngày) với những sản phẩm có cân nặng dưới 3kg và có thể tích nhỏ (<500 cm3) 

image.png

Giả thiết các khách hàng ở nước ngoài (ngoài Brazil) hoặc ở các khu vực rừng Amazon sẽ gây khó khăn hơn co việc vận chuyển dẫn đến kéo dài thời gian giao hàng. Tuy nhiên, Bubble map cho thấy những thành phố của khách hàng ghi nhận trung bình thời gian giao hàng chậm hơn những nơi khác (>1400 giờ) rải rác ở khắp các thành phố của Brazil. Các thành phố lớn và ghi nhận nhiều đơn đặt hàng nhất như Sao Paolo, Rio de Jeneiro cho thấy trung bình thời gian giao hàng tốt hơn
